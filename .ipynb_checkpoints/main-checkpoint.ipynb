```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
car_noise_path='car_noise_specification_datasets.csv'
temp=pd.read_csv(car_noise_path,sep='\t',engine='python')
temp1=temp.iloc[:,range(1,11)].copy()
noise_df=temp1.dropna(how='any')
noise_df.drop_duplicates(subset=['model'])
columns=temp1.columns
temp1
```

```python
noise_df=noise_df.drop_duplicates(subset=['model'],ignore_index=True)
print(columns)
noise_df.head(20)
```

```python
for num in range(len(noise_df)):
    noise_df.iloc[num,0]=noise_df.iloc[num,0]+'_'+noise_df.iloc[num,1]
noise_df=noise_df.drop(['model'],axis=1)
noise_df=noise_df.rename(columns={'brand':'Manufacturer_Model'})
noise_df=noise_df.set_index('Manufacturer_Model')
```

```python
noise_df
```

```python
for col in noise_df.columns:
    print(col+'-----'+str(type(noise_df[col][0])))
```

```python
noise_df['year']=noise_df['year'].astype(int)
```

```python
for col in noise_df.columns:
    print(col+'-----'+str(type(noise_df[col][0])))
```

```python
car_sales_path='Car_sales.csv'
temp_sales=pd.DataFrame(pd.read_csv(car_sales_path))
temp_sales2=temp_sales.drop(['__year_resale_value','Power_perf_factor','Wheelbase','Width','Length','Latest_Launch'],axis=1)
car_sales_df=temp_sales2.dropna(axis=0)
for num in range(len(car_sales_df)):
    car_sales_df.iloc[num,0]=car_sales_df.iloc[num,0]+'_'+car_sales_df.iloc[num,1]
car_sales_df=car_sales_df.drop(['Model'],axis=1)
car_sales_df=car_sales_df.rename(columns={'Manufacturer':'Manufacturer_Model'})
car_sales_df=car_sales_df.set_index('Manufacturer_Model')
```

```python
for col in car_sales_df.columns:
    print(col+'-----'+str(type(car_sales_df[col][0])))
```

```python
#car_sales_df['Horsepower']=car_sales_df['Horsepower'].astype(int)
car_sales_df['Fuel_efficiency']=car_sales_df['Fuel_efficiency'].astype(int)

```
